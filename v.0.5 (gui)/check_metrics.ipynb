```python
from deepiction.imagedataset import ImageDataset, TorchDataset
from interactive_kit import imviewer
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
import warnings
import torch
import os
import stackview

dataname, norm, color = 'ctc-glioblastoma', ('minmax', 'none'), False
datapath  = '../datasets/' + dataname

dataset = ImageDataset(datapath, 'train/sources', 'train/targets')
dataset.load(100, norm, color)
dataset.report()

lossname = 'bce'
if lossname == 'bce': 
    lossfunction = torch.nn.BCELoss()
else: 
    lossfunction = torch.nn.MSELoss()

torchdataset = TorchDataset(dataset.sources, dataset.targets, 'cpu', transform=None)
dataloader = DataLoader(torchdataset, batch_size=1, shuffle=False, pin_memory=True)
for i, (preds, targets) in enumerate(dataloader, 0):
    loss = lossfunction(preds, targets)
    print(i, loss.item())
```
